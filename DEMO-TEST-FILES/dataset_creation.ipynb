```python
#generator functions

#variable generator
def variable_generator():
    variable_lst=['i','j','k','sum','difference','var','x','product','quotient','reminder','time','count','flag','my_variable','text_count','num_of_variables','n','num','total','result','res','replace_string','d','e','f']
    for variable in variable_lst[:10]:
        yield variable

#number generator
def number_generator():
    number_lst=['zero','one','two','three','four','five','six','seven','eight','nine','one zero','two zero','two nine','three five','eight six','two four three','four eight five','one zero four','three eight six','seven nine four','eight one nine','one four three eight','six four three two','eight nine four three','two four four eight']
    for number in number_lst[:20]:
        yield number

#operator generator
def operator_generator():
    operators=['equals','not equals','less than','greater than','less than equals','greater than equals','equals equals','plus','minus','slash','slash slash','star','star star','percent','and','or','xor','plus equals','minus equals','star equals','slash equals','and equals','or equals','xor equals']
    for operator in operators:
        yield operator

#boolean operator generator
def boolean_operator_generator():
    operators=['not equals','less than','greater than','less than equals','greater than equals','equals equals']
    for operator in operators:
        yield operator
```

```python
#Replace Functions

#replaces the number in words with corresponding numbers
def number_replacer(string):
    temp = string.replace(' zero ','0').replace(' one ','1').replace(' two ','2').replace(' three ','3').replace(' four ','4').replace(' five ','5').replace(' six ','6').replace(' seven ','7').replace(' eight ','8').replace(' nine ','9')
    temp = temp.replace('zero ','0').replace('one ','1').replace('two ','2').replace('three ','3').replace('four ','4').replace('five ','5').replace('six ','6').replace('seven ','7').replace('eight ','8').replace('nine ','9')
    temp = temp.replace(' zero','0').replace(' one','1').replace(' two','2').replace(' three','3').replace(' four','4').replace(' five','5').replace(' six','6').replace(' seven','7').replace(' eight','8').replace(' nine','9')
    temp = temp.replace('zero','0').replace('one','1').replace('two','2').replace('three','3').replace('four','4').replace('five','5').replace('six','6').replace('seven','7').replace('eight','8').replace('nine','9')
    return temp

#replaces the operators with corresponding operators
def operator_replacer(string):
    temp = string.replace('equals equals','==').replace('less than equals','<=').replace('greater than equals','>=').replace('not equals','!=').replace('plus equals','+=').replace('minus equals','-=').replace('star equals','*=').replace('slash equals','/=').replace('and equals','&=').replace('xor equals','|=').replace('or equals','^=').replace('less than','<').replace('greater than','>').replace('equals','=').replace('plus','+').replace('minus','-').replace('slash slash','//').replace('slash','/').replace('star star','**').replace('star','*').replace('and','&').replace('xor','^').replace('or','|').replace('percent','%').replace('true','True').replace('false','False')
    return temp

def for_loop_replacer(string):
    if 'range of' in string:
        temp = string.replace('range of','range(')
        temp = temp+"):"
    return temp
print(for_loop_replacer(number_replacer('for i in range of one zero')))
```

```python
intent_statements={
    "assignment_expression":[],
    "arithmetic_expression":[],
    "if_statement":["if true","if false"],
    "else_if_statement":["else if true","else if false"],
    "for_loop_declaration":[]
}
```

```python
template_statements={
    "assignment_expression":["$varname equals $number","$varname equals $varname"],
    "arithmetic_expression":["$varname $operator $number","$varname $operator $varname","$number $operator $number","$operator $varname","$operator $number"],
    "if_statement":["if $varname $boolean_operator $varname","if $number $boolean_operator $number","if $varname $boolean_operator $number"],
    "else_if_statement":["else if $varname $boolean_operator $varname","else if $number $boolean_operator $number","else if $varname $boolean_operator $number"],
    "for_loop_declaration":["for $varname in range of $number","for $varname in range of $varname","for $varname in $varname"]
}
```

```python
for key in template_statements:
    if key=='assignment_expression':
        for statement in template_statements[key]:
            for variable in variable_generator():
                for number in number_generator():
                    temp = statement.replace('$varname',variable)
                    temp = temp.replace('$number',number)
                    intent_statements['assignment_expression'].append(temp)
    elif key=='arithmetic_expression':
        for statement in template_statements[key]:
            for variable in variable_generator():
                for number in number_generator():
                    for operator in operator_generator():
                        temp = statement.replace('$varname',variable)
                        temp = temp.replace('$number',number)
                        temp = temp.replace('$operator',operator)
                        intent_statements['arithmetic_expression'].append(temp)
    elif key=='if_statement':
        for statement in template_statements[key]:
            for variable in variable_generator():
                for number in number_generator():
                    for operator in boolean_operator_generator():
                        temp = statement.replace('$varname',variable)
                        temp = temp.replace('$number',number)
                        temp = temp.replace('$boolean_operator',operator)
                        intent_statements['if_statement'].append(temp)
    elif key=='else_if_statement':
        for statement in template_statements[key]:
            for variable in variable_generator():
                for number in number_generator():
                    for operator in boolean_operator_generator():
                        temp = statement.replace('$varname',variable)
                        temp = temp.replace('$number',number)
                        temp = temp.replace('$boolean_operator',operator)
                        intent_statements['else_if_statement'].append(temp)
    elif key=='for_loop_declaration':
        for statement in template_statements[key]:
            for variable in variable_generator():
                for number in number_generator():
                    temp = statement.replace('$varname',variable)
                    temp = temp.replace('$number',number)
                    intent_statements['for_loop_declaration'].append(temp)
print(len(intent_statements['assignment_expression']))
print(len(intent_statements['arithmetic_expression']))
print(len(intent_statements['if_statement']))
print(len(intent_statements['else_if_statement']))
print(len(intent_statements['for_loop_declaration']))
```

```python
default_file_location = 'D:\\7th Semester\\PROJECT-WORK-PHASE-1\\CODEX-PROJECT-REPO\\AUDIO-FILES'
```

```python
def save_audio(audio_string):
    tts = gTTS(text=audio_string,lang='en')
    r = random.randint(1,1000000000)
    audio_file = 'audio-'+str(r)+'.mp3'
    tts.save(os.path.join(default_file_location,audio_file))
    return audio_file
```

```python
dataset={
    "english_statement":[],
    "intent":[],
    "programming_statement":[],
    "audio_file_name":[]
}
```

```python
#Dataset Creation
def create_dataset(key):
    n=len(intent_statements[key])
    i=1
    for statement in intent_statements[key][:10]:
            temp = operator_replacer(number_replacer(statement))
            if key=='if_statement':
                temp=temp+":"
            elif key=="else_if_statement":
                temp =temp.replace('else if','elsif')
                temp = temp+":"
            elif key=='for_loop_declaration':
                temp = for_loop_replacer(number_replacer(statement))
            dataset['english_statement'].append(statement)
            dataset['intent'].append(key)
            dataset['programming_statement'].append(temp)
            dataset['audio_file_name'].append(save_audio(statement))
            print(f"{i}/{n} completed")
            i+=1
```
